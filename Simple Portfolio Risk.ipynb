```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as ksr
plt.style.use('fivethirtyeight')
```

```python
stocks = ['FB','AMZN','AAPL','TSLA']
```

```python
prices = ksr.DataReader(stocks, data_source= 'yahoo',start = '2019-01-01')['Adj Close']
```

```python
#Log Returns
returns = np.log(prices/prices.shift(1))
```

```python
(prices/prices.iloc[0]*100).plot()
```

```python
returns[['FB','AMZN']].std() * 250 ** 0.5
```

```python
weights = np.array([0.1,0.5,0.2,0.2])
```

```python
#Using Matix Multiplication dot Meathod, (aB)^2 = a.T*B*a
port_var = (np.dot(weights.T,np.dot(returns.cov()*250,weights)))**0.5
print((str(port_var * 100)) + '%')
```
